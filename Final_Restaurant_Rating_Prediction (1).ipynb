```python

from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import r2_score
```

```python

X= dataframe.drop('Aggregate rating', axis=1)
y = dataframe['Aggregate rating']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.1,random_state=353)
X_train.head()
y_train.head()
```

```python

regre=LinearRegression()
regre.fit(X_train,y_train)
y_pred=regre.predict(X_test)
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
regre = LinearRegression()
regre.fit(X_train, y_train)
y_pred = regre.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error (MSE): {mse:.2f}")
r2 = r2_score(y_test, y_pred)
print(f"R-squared (R2) Error: {r2:.2f}")
```

```python
from sklearn.tree import DecisionTreeRegressor
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=10,random_state=105)
DTree=DecisionTreeRegressor(min_samples_leaf=.0001,max_depth=5)
DTree.fit(X_train,y_train)
y_predict=DTree.predict(X_test)  
from sklearn.metrics import r2_score
```

```python
# Calculate MSE
mse = mean_squared_error(y_test, y_predict)
print("Mean Squared Error:", mse)
```

```python
DTree.score(X_train,y_train)
```

```python
DTree.score(X_test,y_test)
```
